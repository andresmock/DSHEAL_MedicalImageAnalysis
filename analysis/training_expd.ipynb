```python
import os
import wandb
import pandas as pd
```

```python
login_file = ".wandb_login"

if os.path.exists(login_file):
    with open(login_file, "r") as f:
        for line in f:
            if line.startswith("WANDB_API_KEY="):
                os.environ["WANDB_API_KEY"] = line.strip().split("=")[1]
                break

wandb.login()
```

```python
api = wandb.Api()

# Dein W&B-Projekt auswählen
project_name = "mockand1-ba/DSHEAL_pro1_FS25_GaMo"

# Alle gespeicherten Runs abrufen
runs = api.runs(project_name)
```

```python
# Runs in ein DataFrame konvertieren
history = []

for run in runs:
    df = run.history(keys=["Epoch", "Train Accuracy", "Validation Accuracy", "Train Loss (Epoch Avg)", "Validation Loss"])
    df["run_id"] = run.id
    df["run_name"] = run.name
    history.append(df)

# Alle Runs zusammenfügen
df_wandb = pd.concat(history, ignore_index=True)
df_wandb
```

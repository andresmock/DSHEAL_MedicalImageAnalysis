```python
import optuna
import pandas as pd

study = optuna.load_study(
    study_name="malaria_phase1_basics",
    storage="sqlite:///C:/Users/andre/Desktop/git_repository_zhaw/DSHEAL_project_FS25_GaMo/malaria_optuna.db"
)


data = []
for trial in study.trials:
    row = {
        "Trial": trial.number,
        "Status": trial.state.name,  # Status als lesbaren Namen
        "Value": trial.value
    }
    row.update(trial.params)  # Parameter hinzufügen
    data.append(row)

df = pd.DataFrame(data)
df_sorted = df.sort_values(by="Value", ascending=False)
df_sorted
```
